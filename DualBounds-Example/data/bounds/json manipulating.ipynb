Combining data from jsons

```python
import json

def merge_dicts(dict1, dict2):
    """
    Recursively merges two dictionaries.
    """
    merged = dict1.copy()
    for key, value in dict2.items():
        if key in merged and isinstance(merged[key], dict) and isinstance(value, dict):
            merged[key] = merge_dicts(merged[key], value)
        else:
            merged[key] = value
    return merged

# File paths
state = 'MS'
file1_path = f'{state}_partisan_bounds.json'
file2_path = f'{state}_partisan_bounds - Copy.json'
file2_path = f'{state}_partisan_bounds - combined.json'
combined_file_path = f'{state}_partisan_bounds - combined.json'

# Load the first JSON file
with open(file1_path, 'r') as file1:
    data1 = json.load(file1)

# Load the second JSON file
with open(file2_path, 'r') as file2:
    data2 = json.load(file2)

# Combine the dictionaries
combined_data = merge_dicts(data1, data2)

# Save the combined data as a new JSON file
with open(combined_file_path, 'w') as outfile:
    json.dump(combined_data, outfile, indent=4)
```

```python
import json

# Load the JSON data
state = 'MS'
file_path = f'{state}_partisan_bounds - combined.json'
outfile_path = f'{state}_partisan_bounds - formatted.json'
with open(file_path, 'r') as file:
    data = json.load(file)

# Initialize an empty dictionary for the reformatted data
new_data = {}

# List of variables
variables = ["D20", "D16", "R20", "R16", "T20", "T16", "diff_D_T20", "diff_D_T16"]

# Iterate over each variable
for var in variables:
    # Initialize dictionaries for lower and upper bounds
    lb_dict = {}
    ub_dict = {}
    min_soln_dict = {}
    max_soln_dict = {}

    # Iterate over each key in the original data
    for key in data['Bounds']:
        # Check if the key starts with the current variable
        if key.startswith(var):
            # Extract the district index
            index = key[len(var):]

            # Add the lower and upper bounds to the corresponding dictionaries
            lb_dict[index] = data['Bounds'][key]['lb']
            ub_dict[index] = data['Bounds'][key]['ub']
            min_soln_dict[index] = data['Bounds'][key]["min_soln"]
            max_soln_dict[index] = data['Bounds'][key]["min_soln"]

    # Add the lower and upper bounds to the new data
    new_data[var] = {'lb': lb_dict, 'ub': ub_dict, 'min_soln':min_soln_dict, 'max_soln':max_soln_dict }

# Print the new data
print(json.dumps(new_data, indent=2))

# Save the new data to a JSON file
with open(outfile_path, 'w') as file:
    json.dump(new_data, file, indent=4)
```
