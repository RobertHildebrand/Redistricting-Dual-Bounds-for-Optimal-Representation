```python
import json
import matplotlib.pyplot as plt

def plot_for_state(state, ax):
    # Specify the path to your JSON file
    json_file_path = f"../data/bounds/{state}_bvap_gradient_bounds.json"

    # Load the JSON data from the file
    with open(json_file_path, "r") as json_file:
        data = json.load(json_file)

    # Extract the 'ub' values for each integer key
    keys = sorted(data.keys(), key=lambda x: int(x))  # Sort keys in ascending order
    ub_values = [data[key]['Relaxation_bound'] for key in keys]
    min_bounds = [min(ub_values[:i]) for i in range(1, len(ub_values) + 1)]

    # Create a plot for the current state
    ax.plot(keys, min_bounds)
    ax.set_xlabel('Keys')
    ax.set_ylabel('Relaxation bound')
    ax.set_title(f'Bounds after each gradient cut ({state})')

# Create a figure and call the plot function for each state
fig, axes = plt.subplots(2, 2, figsize=[10, 8])  # Adjust the figsize based on your preferences

# Specify different figsize values for each subplot
subplots = {
    'MS': {'subplot': axes[0, 0], 'figsize': [6, 4]},
    'SC': {'subplot': axes[0, 1], 'figsize': [8, 6]},
    'AL': {'subplot': axes[1, 0], 'figsize': [20, 4]},
    'LA': {'subplot': axes[1, 1], 'figsize': [8, 8]}
}

# Call the plot_for_state() function for each state and set the corresponding figsize
for state, subplot in subplots.items():
    plot_for_state(state, subplot['subplot'])
    subplot['subplot'].figure.set_size_inches(subplot['figsize'])

# Adjust the layout of subplots
fig.tight_layout()

# Display the plot
plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import pandas as pd

def plot_for_state(state):
    # Specify the path to your JSON file
    json_file_path = f"../data/bounds/{state}_bvap_gradient_bounds.json"
    csv_file_path = f"../results/objective_sorted/{state}_bvap_sorted.csv"

    # Load the JSON data from the file
    with open(json_file_path, "r") as json_file:
        data = json.load(json_file)

    # Extract the 'ub' values and 'ub_computation (s)' for each integer key
    keys = sorted(data.keys(), key=lambda x: int(x))  # Sort keys in ascending order
    ub_values = [data[key]['Relaxation_bound'] for key in keys]
    ub_computation = [data[key]['ub_computation (s)'] for key in keys]
    min_bounds = [min(ub_values[:i]) for i in range(1, len(ub_values) + 1)]

    # Load the CSV file and extract the desired value
    df = pd.read_csv(csv_file_path)
    target_value = df.iloc[0, 2]  # Second column, first row after the header

    # Create a subplot for the current state
    fig, ax1 = plt.subplots(figsize=(8, 6))  # Adjust the figsize as desired

    # Plot 'min_bounds' with the primary y-axis
    ax1.plot(keys, min_bounds, color='blue')
    ax1.axhline(target_value, color='r', linestyle='--')  # Plot dashed line for target_value
    ax1.set_xlabel('Keys')
    ax1.set_ylabel('Relaxation bound')
    ax1.set_title(f'Bounds after each gradient cut ({state})')

    # Create a secondary y-axis for 'ub_computation' on the right side
    ax2 = ax1.twinx()
    ax2.plot(keys, ub_computation, color='green', linestyle='--')
    ax2.set_ylabel('ub_computation (s)')

    # Adjust the limits of the primary and secondary y-axes
    #ax1.set_ylim(0, max(min_bounds)*1.1)
    ax2.set_ylim(0, max(ub_computation)*1.1)

    # Show the legend for both lines on the plot
    ax1.legend(['Relaxation bound', 'Target value'], loc='upper left')
    ax2.legend(['ub_computation'], loc='upper right')

    # Display the plot
    plt.show()
    
# Create a figure and call the plot function for each state
fig = plt.figure(figsize=(14, 8))  # Adjust the figsize to accommodate the widest plot

# Call the plot function for each state
# Subplot for 'MS'
plt.subplot(2, 2, 1)
plot_for_state('MS')

# Subplot for 'SC'
plt.subplot(2, 2, 2)
plot_for_state('SC')

# Subplot for 'AL'
plt.subplot(2, 2, 3)
plot_for_state('AL')

# Subplot for 'LA'
plt.subplot(2, 2, 4)
plot_for_state('LA')

plt.show()
```

```python
import json
import matplotlib.pyplot as plt
import pandas as pd

def plot_for_state(state, ax):
    # Specify the path to your JSON file
    json_file_path = f"../data/bounds/{state}_bvap_gradient_bounds.json"
    csv_file_path = f"../results/objective_sorted/{state}_bvap_sorted.csv"

    # Load the JSON data from the file
    with open(json_file_path, "r") as json_file:
        data = json.load(json_file)

    # Extract the 'ub' values and 'ub_computation (s)' for each integer key
    keys = sorted(data.keys(), key=lambda x: int(x))  # Sort keys in ascending order
    ub_values = [data[key]['Relaxation_bound'] for key in keys]
    ub_computation = [data[key]['ub_computation (s)'] for key in keys]
    min_bounds = [min(ub_values[:i]) for i in range(1, len(ub_values) + 1)]

    # Load the CSV file and extract the desired value
    df = pd.read_csv(csv_file_path)
    target_value = df.iloc[0, 2]  # Second column, first row after the header

    # Create a plot for the current state
    ax.plot(keys, min_bounds, color='blue')
    ax.axhline(target_value, color='r', linestyle='--')  # Plot dashed line for target_value
    ax.set_xlabel('Keys')
    ax.set_ylabel('Relaxation bound')
    ax.set_title(f'Bounds after each gradient cut ({state})')

    # Create a secondary y-axis for 'ub_computation' on the right side
    ax2 = ax.twinx()
    ax2.plot(keys, ub_computation, color='green', linestyle='--')
    ax2.set_ylabel('ub_computation (s)')

    # Adjust the limits of the primary and secondary y-axes
    #ax.set_ylim(0, max(min_bounds)*1.1)
    ax2.set_ylim(0, max(ub_computation[0:min(30, len(keys)-1)])*1.1)
    ax.set_xlim(0,min(30, len(keys)-1))

    # Show the legend for both lines on the plot
    ax.legend(['Relaxation bound', 'Target value'], loc='upper left')
    ax2.legend(['time spent for each cut'], loc='upper right')

# Create a figure and call the plot function for each state
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Adjust the figsize as desired

# Subplot for 'MS'
plot_for_state('MS', axes[0, 0])

# Subplot for 'SC'
plot_for_state('SC', axes[0, 1])

# Subplot for 'AL'
plot_for_state('AL', axes[1, 0])

# Subplot for 'LA'
plot_for_state('LA', axes[1, 1])

# Adjust the layout of subplots
fig.tight_layout()

plt.savefig('gradient_cuts_with_time.png')
# Display the plot
plt.show()
```
