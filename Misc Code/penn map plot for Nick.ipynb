```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
penn = gpd.read_file('Penn_map/BlockLevelFinalCongressionalPlan21Dec2011.shp')
```

```python

penn.plot(column = 'OBJECTID')
```

```python
ax = penn.plot(column = 'OBJECTID')
ax.set_axis_off();
plt.savefig('penn.eps',  transparent=True)
```

```python



ax = penn.plot(column = 'OBJECTID')
ax.set_axis_off();
plt.savefig('penn.svg',  transparent=True)
```

```python
ax = penn.plot(column = 'OBJECTID')
ax.set_axis_off();
plt.savefig('penn.png',  transparent=True)
```

```python
ax = penn.plot(column = 'OBJECTID')
ax.set_axis_off();
plt.savefig('penn.pdf',  transparent=True)
```

```python
for i in penn['OBJECTID']:
    district = penn[penn['OBJECTID']==i]
    ax = district.plot()
    ax.set_axis_off();
    plt.title(f'2012 Penn District {i}')
    plt.savefig(f'penn_2012_district_{i}.png',  transparent=True)
```
